   **Expicação do código em baixo**

1. No código estamos a criar uma lista ```var1```, para armazenar os dados;
2. Gerar dados (números aleatórios) ```random.seed(24)```, mas que sejam sempre os mesmos 24 números a serem gerados;
3. Definir as variáveis ```levels``` (categorias de uso de terra);
4. Gerar 100 valores aleatórios para cada categoria ```for_in range(100)```;
5. Escolher n elementos da lista, sem repetir em cada loop (```random.sample(levels, 1)```). Mas ao longo do tempo, visto que irá correr 100 vezes o loop, os valores podem repetir;
6. Adicionar esses valores à lista ```var1 + = random.sample(levels, 1)```.

````python
# Simulate var1
var1 = []
random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]
for _ in range(100): # a loop is needed because random.sample selects unique elements (with no replacement)
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work
print(var1)
````

   ### **Alternativa para executar uma amostragem aleatória com substituição (usando ```numpy```)**

1. ```levels``` é um ```array NumPy``` que contém 8 categorias diferentes de ocupação do solo, como culturas permanentes, florestas, pastagens, entre outras. Este ```array``` é usado como base para selecionar amostras aleatórias;
2. ```np.random.randint(0, len(levels), 100)``` gera 100 números inteiros aleatórios entre 0 e 7 (porque ```len(levels) = 8```). Como usamos ```randint```, os números podem repetir-se, o que significa que estamos a fazer amostragem com reposição (um elemento pode ser selecionado mais do que uma vez). O resultado será um ```array``` com 100 números inteiros;
3. ```levels.take(sampler)``` usa os valores gerados em ```sampler``` para selecionar os elementos correspondentes do array ```levels```;
4. ```print(var1)``` exibe no ecrã a lista de 100 elementos selecionados aleatoriamente de ```levels```.

````python
# alternative to run a random sampling with replacement (using numpy)
levels = np.array(["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland"])
sampler = np.random.randint(0, len(levels), 100) # 100 random values within an interval (0 to 7)
var1 = levels.take(sampler) # use sampler to select values from "levels"; take - returns elements from array along the mentioned axis and indices
print(var1)
````

**Explicação extra: Qual a diferença entre ```numpy.ndarray``` e ```list```?**

A diferença principal entre um NumPy array (```numpy.ndarray```) e uma lista (```list```) do Python é a eficiência:

- Arrays NumPy são **mais rápidos** e ocupam **menos memória**;
- Arrays NumPy permitem **operações matemáticas** diretamente (ex: ```sampler * 2``` multiplica todos os elementos);
- Listas Python são mais **flexíveis**, mas menos eficientes para cálculos numéricos.

**Explicação do próximo código**

1. ```np.random.seed(número)``` define uma **seed** (semente) para o gerador de números aleatórios. Isto faz com que o código gere **sempre os mesmos números** ao ser executado novamente. Se não usares ```np.random.seed()```, os valores serão diferentes a cada execução;
2. ```np.random.uniform(a, b, n)``` gera ```n``` números aleatórios de uma distribuição **uniforme** entre ```a``` e ```b```. Neste caso, gera **100 números** decimais entre ```0``` e ```100```. Como é uma distribuição **uniforme**, qualquer número dentro do intervalo tem a mesma probabilidade de ser escolhido;
3. ```print(var2)```mostra os 100 valores gerados.


````python
# Simulate var2
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
var2 = np.random.uniform(0, 100, 100)
print(var2)
````

## **Criar tabelas**

**1. Criar um DataFrame a partir de ```var1```**

- ```pd.DataFrame(var1)``` cria um **DataFrame** do Pandas a partir do array ```var1```;
- ```value_counts(sort=True)``` conta quantas vezes cada tipo de uso do solo aparece e ordena os resultados por frequência (por padrão, da maior para a menor);
- Neste momento, ```table1``` ainda **não é um DataFrame normal**, mas sim uma ```Series``` com índices nomeados pelos valores únicos de ```var1```.

**2. Renomear o índice para "landuse"**

- ```rename_axis("landuse")``` dá um nome ao índice (antes era apenas um índice anónimo);
- Agora os valores categóricos (ex: "Urban", "Pasture") têm um nome "**landuse**".

**3. Transformar ```table1``` num DataFrame normal**

- ```reset_index(name="Frequency")``` transforma o índice em **coluna** e dá-lhe o nome "**landuse**";
- O que antes era a contagem passa a ser a coluna ```Frequency```.

**4. Imprimir a tabela**

- ```print(table1)```Agora ```table1``` é um **DataFrame organizado**, onde cada tipo de ocupação do solo (```landuse```) tem a sua frequência (```Frequency```).

````python
# Simulate table1
table1 = pd.DataFrame(var1).value_counts(sort=True)
table1 = table1.rename_axis("landuse")
table1 = table1.reset_index(name="Frequency")
print(table1)
````

````python
# Simulate table4
xx = np.array([16,21])
yy = np.array([300, 1200])
means = [xx.mean(), yy.mean()]  
stds = [xx.std() / 3, yy.std() / 3]
corr = -0.7 # correlation
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])
print(table4)
````
